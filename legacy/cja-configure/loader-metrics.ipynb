```python
import pandas as pd
import re
import os
from openpyxl import load_workbook

# 1. The Raw Input Data
raw_input = input

def generate_xdm_path(event_num):
    """
    Maps event numbers to the correct Adobe Analytics XDM path.
    Events 1-100 live in 'event1to100', 101-200 in 'event101to200', etc.
    """
    n = int(event_num)
    if 1 <= n <= 100:
        group = "event1to100"
    elif 101 <= n <= 200:
        group = "event101to200"
    elif 201 <= n <= 300:
        group = "event201to300"
    elif 301 <= n <= 400:
        group = "event301to400"
    elif 401 <= n <= 500:
        group = "event401to500"
    else:
        group = "event501to600" # rare but possible
        
    return f"_experience.analytics.{group}.event{n}.value"

def parse_and_save_metrics(input_text, filename):
    rows = []
    
    # --- A. Parse Text ---
    for line in input_text.strip().split('\n'):
        if not line.strip(): continue
        
        # Extract Event Number (ID)
        parts = line.split(': ', 1)
        event_label = parts[0]
        remainder = parts[1] if len(parts) > 1 else ""
        
        match_num = re.search(r'Custom Event (\d+)', event_label)
        event_num = match_num.group(1) if match_num else "0"
        
        # Calculate XDM Path
        xdm_path = generate_xdm_path(event_num)
        
        # Extract attributes
        name = ""
        desc = ""
        dedup = ""
        metric_type = ""
        
        attributes = remainder.split(': ')
        for attr in attributes:
            if '=' in attr:
                key, value = attr.split('=', 1)
                key = key.strip()
                value = value.strip()
                
                if key == "Name":
                    #name = re.sub(r'\s\[e\d+\]$', '', value) # Remove [e1] suffix
                    name = value
                elif key == "Type":
                    metric_type = value
                elif key == "Description":
                    desc = value
                elif key == "Unique Event Record":
                    dedup = value

        # Map "Unique Event Record" to CJA Deduplication setting
        # CJA values: "session" (Visit), "person" (Visitor), or empty (Always Record)
        dedup_setting = ""
        if "Visit" in dedup:
            dedup_setting = "session"
        elif "Visitor" in dedup:
            dedup_setting = "person"
            
        rows.append({
            "component_id": "",         # Empty -> To be filled by sync
            "status": "Disabled" if metric_type == "Disabled" else "Pending",
            "name": name,
            "description": desc,
            "id": f"event{event_num}",  # e.g., "event1"
            "source": xdm_path,
            "deduplication": dedup_setting,
            "type": metric_type         # Helpful for reference
        })
    
    df_metrics = pd.DataFrame(rows)
    
    # --- B. Save to Excel (Append Mode) ---
    try:
        # Check if file exists first
        if not os.path.exists(filename):
            # Create new if doesn't exist
            with pd.ExcelWriter(filename, engine='openpyxl') as writer:
                df_metrics.to_excel(writer, sheet_name='metrics', index=False)
        else:
            # Append to existing file without overwriting 'dimensions' sheet
            with pd.ExcelWriter(filename, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
                df_metrics.to_excel(writer, sheet_name='metrics', index=False)
                
        print(f"Success! Added {len(df_metrics)} metrics to sheet 'metrics' in '{filename}'.")
        return df_metrics
        
    except PermissionError:
        print(f"ERROR: Please close '{filename}' and try again.")
    except Exception as e:
        print(f"An error occurred: {e}")

# --- Run It ---
output_file = "cja-metrics-dimensions.xlsx"
df_result = parse_and_save_metrics(raw_input, output_file)

if df_result is not None:
    print(df_result[['id', 'name', 'source', 'deduplication']].head())
```
